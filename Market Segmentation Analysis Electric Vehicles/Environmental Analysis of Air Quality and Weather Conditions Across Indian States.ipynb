```python
import pandas as pd 
import numpy as np                     # For mathematical calculations 
import seaborn as sns                  # For data visualization 
import matplotlib.pyplot as plt        # For plotting graphs 
%matplotlib inline 
import warnings   # To ignore any warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("pollution data")
df
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='AQI-US', y='States', data=df, palette='viridis')
plt.title('Air Quality Index (AQI) Across States in India')
plt.xlabel('AQI-US')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(y='Status', data=df, palette='Set2', order=df['Status'].value_counts().index)
plt.title('Distribution of Air Quality Status')
plt.xlabel('Count')
plt.ylabel('Air Quality Status')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temp', y='AQI-US', data=df, hue='Status', palette='coolwarm', size='Humid', sizes=(30, 200))
plt.title('Temperature vs AQI-US')
plt.xlabel('Temperature (°C)')
plt.ylabel('AQI-US')
plt.legend(title='Air Quality Status', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data (replace with actual data from your analysis)
cities = ['Delhi', 'Mumbai', 'Bengaluru', 'Chennai']
aqi_values = [108, 62, 40, 61]  # AQI values
charging_stations = [500, 300, 250, 200]  # Number of EV charging stations
govt_support = [4, 3, 5, 4]  # Government support score (on a scale of 1 to 5)

# Normalize values for visualization
normalized_aqi = [(aqi - min(aqi_values)) / (max(aqi_values) - min(aqi_values)) * 100 for aqi in aqi_values]
normalized_charging = [(stations - min(charging_stations)) / (max(charging_stations) - min(charging_stations)) * 100 for stations in charging_stations]
normalized_support = [(support / 5) * 100 for support in govt_support]

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.25
index = range(len(cities))

bars1 = ax.bar(index, normalized_aqi, bar_width, label='AQI', color='b', alpha=0.7)
bars2 = ax.bar([i + bar_width for i in index], normalized_charging, bar_width, label='Charging Stations', color='g', alpha=0.7)
bars3 = ax.bar([i + 2 * bar_width for i in index], normalized_support, bar_width, label='Government Support', color='r', alpha=0.7)

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Cities')
ax.set_ylabel('Normalized Scores')
ax.set_title('Comparison of Cities for EV Adoption Readiness')
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(cities)
ax.legend()

# Function to add labels on top of bars
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval + 1, round(yval, 2), ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

plt.tight_layout()
plt.show()
```
